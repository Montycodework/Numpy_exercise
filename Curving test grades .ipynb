```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
CURVE_CENTER = 80
```

```python
grades = np.array([72,35,64,88,51,90,74, 12])
```

```python
grades
```

```python
def curve(grades):
    average = grades.mean()
    change = CURVE_CENTER  - average
    new_grades = grades + change
    return np.clip(new_grades,grades,100) # clip is used to limit the value upto 100
```

```python
new_curve= curve(grades)
new_curve
```

```python
plt.plot(grades)
plt.plot(new_curve)
```
